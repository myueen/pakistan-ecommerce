## Data Cleaning

- Remove the last four columns since they are NaN.

```python
df['Unnamed: 21']
df['Unnamed: 22']
df['Unnamed: 23']
df['Unnamed: 24']
df['Unnamed: 25']
```

```python
df_d = df.drop(columns=['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'])
df_d
```

- Since the dataset consists of about 60 thousands observations that are don't include NaN, set the threshhold to be 584300 and remove the rest that does not include any observation. 

```python
df_no_NaN = df_d.iloc[: 584301, :]
df_no_NaN
```

- In this dataset, we are only interested in the transaction that is completed. 

```python
df_complete = df_no_NaN.loc[df_no_NaN['status'] == "complete"]
df_complete
```

```python
df_complete["created_at"] = pd.to_datetime(df_complete["created_at"] )
df_complete["created_at"] 
```

```python
df_complete['created_at'].dt.year
```

```python
df_complete['created_at'].dt.month
```

```python
df_complete['created_at'].dt.day
```

```python
df_complete["year"] = df_complete['created_at'].dt.year
df_complete["month"] = df_complete['created_at'].dt.month
df_complete["day"] = df_complete['created_at'].dt.day
df_complete
```
